```python
y2020 = [17,13,12,6,5,3,2,42]; y2020_rev = [42, 2, 3, 5, 6, 12, 13, 17]
y2020_label = ["네이버", "쿠팡","이베이코리아","11번가","롯데온","SSG닷컴","카카오","기타"]
y2020_label_rev = ['*기타', '카카오', 'SSG닷컴', '롯데온', '11번가', '이베이코리아', '쿠팡', '네이버']
y2022 = [17,15,13,6,5,4,42]; y2022_rev=[42, 4, 5, 6, 13, 15, 17]
y2022_label = ["네이버", "SSG닷컴(이베이코리아)","쿠팡","11번가","롯데온","카카오","기타"]
y2022_label_rev = ['기타', '카카오', '롯데온', '11번가', '쿠팡', 'SSG닷컴(이베이코리아)', '네이버']
```

```python
plt.figure(figsize=(7,7))
plt.barh(y2022_label_rev, y2022_rev, color="navy")
for i in range(len(y2022)):
    plt.text(y2022[i]+0.2, len(y2022)-i-1.05, str(abs(y2022[i])), fontdict={'color':'navy', 'fontweight':'bold', 'fontsize':13})
plt.title("2022년", fontdict={'fontweight':'bold', 'color':'navy', 'fontsize':20})
plt.yticks(fontsize=13)
```
